# Exercice

y=-5x+9 sur [-20,20]

```python
x=np.arange(-20,20)
y=-5*x+9
print(x)
print(y)
plt.plot(x,y,c='r')
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.title("Fonction y")
plt.legend("y=-5x+9")
plt.show()
```

```python
x= np.linspace(0,8,100)
plt.scatter(x,np.sin(x)+random.rand(x.shape[0])*0.4-0.2)
plt.plot(x,np.sin(x),c='r')
```
